# ***K Means Clustering***

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Machine Learning Class/Data/Mall_Customers.csv')
df
```

```python
df.isnull().sum()
```

```python
x= df.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=1)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,11),wcss)
plt.xlabel(" NO OF CLUSTERS")
plt.ylabel=("WCSS")
plt.title("ELBOW FOR OPTIMAL CLUSTER")
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=1)
kmeans.fit(x)
```

```python
kmeans.cluster_centers_
```

```python
y=kmeans.fit_predict(x)
y
```

```python
plt.scatter(x[y== 0,0],x[y==0,1], c='blue',s=100,label='cluster1',marker='*')
plt.scatter(x[y== 1,0],x[y==1,1], c='yellow',s=100,label='cluster2',marker='D')
plt.scatter(x[y== 2,0],x[y==2,1], c='green',s=100,label='cluster3',marker='^')
plt.scatter(x[y== 3,0],x[y==3,1], c='red',s=100,label='cluster4',marker='s')
plt.scatter(x[y== 4,0],x[y==4,1], c='cyan',s=100,label='cluster5',marker='p')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=200,c='black',label="centroid")
plt.legend()
plt.xlabel(" annual income")
plt.ylabel=("spending score")
plt.title("customerÂ analysis")
```
